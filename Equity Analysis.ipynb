**TIME SERIES DATA**
A time series is a sequence of numerical data points taken at successive equally spaced points in time.
Here, a time series tracks the movement of stock price, over a specified period of time.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates
```

```python
plt.plot(msft.index, msft['Adj Close'])
plt.grid(True)
plt.xticks(rotation=0)
plt.show()
```

**WEEKLY RESAMPLING**

```python
msft_2014 = msft.loc[pd.Timestamp('2014-01-01'):pd.Timestamp('2014-08-31')]
weekly_msft_2014 = msft_2014.resample('W').mean()
weekly_msft_2014.head()
```

```python
plt.plot(weekly_msft_2014.index, weekly_msft_2014['Adj Close'], '-o')
```

```python
nvda_14 = nvda.loc[pd.Timestamp('2019-01-15'):pd.Timestamp('2019-09-15')]
weekly_nvda_14 = nvda_14.resample('W').mean()

intc_14 = intc.loc[pd.Timestamp('2019-01-15'):pd.Timestamp('2019-09-15')]
weekly_intc_14 = intc_14.resample('W').mean()

qcom_14 = qcom.loc[pd.Timestamp('2019-01-15'):pd.Timestamp('2019-09-15')]
weekly_qcom_14 = qcom_14.resample('W').mean()

msft_14 = msft.loc[pd.Timestamp('2019-01-15'):pd.Timestamp('2019-09-15')]
weekly_msft_14 = msft_14.resample('W').mean()

f, ax = plt.subplots(2, 2, figsize=(10,10), sharex=True, sharey=True)
ax[0,0].plot(weekly_nvda_14.index, weekly_nvda_14['Adj Close'], '-o', color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('NVIDIA');

ax[0,1].plot(weekly_intc_14.index, weekly_intc_14['Adj Close'], '-o',color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('INTEL');

ax[1,0].plot(weekly_qcom_14.index, weekly_qcom_14['Adj Close'],'-o', color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('QUALCOMM');

ax[1,1].plot(weekly_msft_14.index, weekly_msft_14['Adj Close'], '-o', color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('MSFT');
```

**MOVING WINDOWS** - **Daily percentages**

```python
daily_close_msft = msft[['Adj Close']]
#Daily returns
daily_pct_change_msft = daily_close_msft.pct_change()

# Clean data: Replace null values with 0
daily_pct_change_msft.fillna(0,inplace=True)
daily_pct_change_msft.head()
```

```python
daily_pct_change_msft.hist(bins=100)
plt.show()
```

```python
daily_close_nvda = nvda[['Adj Close']]

# Daily returns
daily_pct_change_nvda = daily_close_nvda.pct_change()

# Replace NA values with 0
daily_pct_change_nvda.fillna(0, inplace=True)

daily_close_intc = intc[['Adj Close']]

# Daily returns
daily_pct_change_intc = daily_close_intc.pct_change()

# Replace NA values with 0
daily_pct_change_intc.fillna(0, inplace=True)

daily_close_qcom = qcom[['Adj Close']]

# Daily returns
daily_pct_change_qcom = daily_close_qcom.pct_change()

# Replace NA values with 0
daily_pct_change_qcom.fillna(0, inplace=True)

daily_close_msft = msft[['Adj Close']]

# Daily returns
daily_pct_change_msft = daily_close_msft.pct_change()

# Replace NA values with 0
daily_pct_change_msft.fillna(0, inplace=True)
```

```python

import seaborn as sns
# Set up the matplotlib figure
f, axes = plt.subplots(2, 2, figsize=(12, 7))

sns.distplot(daily_pct_change_nvda['Adj Close'], color="b", ax=axes[0, 0], axlabel='NVIDIA');
sns.distplot(daily_pct_change_intc['Adj Close'], color="r", ax=axes[0, 1], axlabel='INTEL');
sns.distplot(daily_pct_change_qcom['Adj Close'], color="g", ax=axes[1, 0], axlabel='QUALCOMM');
sns.distplot(daily_pct_change_msft['Adj Close'], color="m", ax=axes[1, 1], axlabel='MSFT');
```

**VOLATALITY PRESENCE**

```python
import numpy as np 
min_periods = 75
#Calculating the volatility 
vol = daily_pct_change_msft.rolling(min_periods).std()*np.sqrt(min_periods)
vol.fillna(0,inplace = True)
vol.tail()
```

```python
# pictorial representation of the volatility in MSFT
vol.plot(figsize=(10,8))
plt.show()
```

**TRENDS AND SEASONALITY**

```python
ibm_adj_close_price = ibm['Adj Close']
# Short moving window rolling mean
ibm['42'] = ibm_adj_close_price.rolling(window=42).mean()

# Long moving window rolling mean
ibm['252'] = ibm_adj_close_price.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
ibm[['Adj Close', '42', '252']].plot(title="IBM")

# Show plot
plt.show()

nvda_adj_close_price = nvda['Adj Close']
# Short moving window rolling mean
nvda['42'] = nvda_adj_close_price.rolling(window=42).mean()

# Long moving window rolling mean
nvda['252'] = nvda_adj_close_price.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
nvda[['Adj Close', '42', '252']].plot(title="NVIDIA")

# Show plot
plt.show()
```
